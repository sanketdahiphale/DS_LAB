```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)

# Example usage:
number = 5
print(f"The factorial of {number} is {factorial(number)}")

number = 0
print(f"The factorial of {number} is {factorial(number)}")

number = 7
print(f"The factorial of {number} is {factorial(number)}")
```

```python
def check_odd_even(number):
    if number % 2 == 0:
        print(f"The number {number} is Even.")
    else:
        print(f"The number {number} is Odd.")

# Example usage:
check_odd_even(4)
check_odd_even(7)
check_odd_even(0)
check_odd_even(15)
```

```python
def is_palindrome(number):
    # Convert the number to a string to easily reverse it
    num_str = str(number)
    # Compare the string with its reverse
    if num_str == num_str[::-1]:
        print(f"The number {number} is a palindrome.")
    else:
        print(f"The number {number} is not a palindrome.")

# Example usage:
is_palindrome(121)
is_palindrome(12321)
is_palindrome(123)
is_palindrome(5)
is_palindrome(0)
```

```python
rules = [
    {
        "conditions": {
            "weather": "sunny",
            "temperature": "warm"
        },
        "recommendation": "Go for a walk or a picnic in the park."
    },
    {
        "conditions": {
            "weather": "rainy"
        },
        "recommendation": "Read a book, watch a movie, or play board games indoors."
    },
    {
        "conditions": {
            "weather": "cold",
            "temperature": "freezing"
        },
        "recommendation": "Stay indoors, have hot chocolate, or go ice skating if possible."
    },
    {
        "conditions": {
            "weather": "cloudy",
            "temperature": "mild"
        },
        "recommendation": "Visit a museum or art gallery."
    }
]

print("Expert system rules defined:")
for i, rule in enumerate(rules):
    print(f"Rule {i+1}:")
    print(f"  Conditions: {rule['conditions']}")
    print(f"  Recommendation: {rule['recommendation']}")
```

## Implement Expert System

### Subtask:
Write a Python function or class that takes relevant inputs (facts) and applies the defined rules to deduce a decision or recommendation. This will involve implementing a basic inference engine.


**Reasoning**:
Implement the `recommend_activity` function as described in the instructions, which will act as the basic inference engine for the expert system by matching input facts against the predefined rules.



```python
def recommend_activity(facts):
    for rule in rules:
        # Assume the rule matches until proven otherwise
        rule_matches = True
        for condition_key, condition_value in rule['conditions'].items():
            # Check if the fact exists and matches the condition
            if condition_key not in facts or facts[condition_key] != condition_value:
                rule_matches = False
                break  # No need to check other conditions for this rule

        if rule_matches:
            return rule['recommendation']

    return "No specific recommendation based on current facts."

print("Inference engine function 'recommend_activity' defined.")
```

**Reasoning**:
The `recommend_activity` function has been defined. Now, demonstrate its usage with example `facts` to show how the expert system provides recommendations based on the rules.



```python
print("\n--- Testing the Expert System ---")

# Test Case 1: Sunny and warm
current_facts_1 = {'weather': 'sunny', 'temperature': 'warm'}
recommendation_1 = recommend_activity(current_facts_1)
print(f"Facts: {current_facts_1} -> Recommendation: {recommendation_1}")

# Test Case 2: Rainy day
current_facts_2 = {'weather': 'rainy'}
recommendation_2 = recommend_activity(current_facts_2)
print(f"Facts: {current_facts_2} -> Recommendation: {recommendation_2}")

# Test Case 3: Cold and freezing
current_facts_3 = {'weather': 'cold', 'temperature': 'freezing'}
recommendation_3 = recommend_activity(current_facts_3)
print(f"Facts: {current_facts_3} -> Recommendation: {recommendation_3}")

# Test Case 4: Cloudy and mild
current_facts_4 = {'weather': 'cloudy', 'temperature': 'mild'}
recommendation_4 = recommend_activity(current_facts_4)
print(f"Facts: {current_facts_4} -> Recommendation: {recommendation_4}")

# Test Case 5: No matching rule (e.g., snowy)
current_facts_5 = {'weather': 'snowy', 'temperature': 'cold'}
recommendation_5 = recommend_activity(current_facts_5)
print(f"Facts: {current_facts_5} -> Recommendation: {recommendation_5}")

# Test Case 6: Partially matching facts (warm but not sunny)
current_facts_6 = {'temperature': 'warm'}
recommendation_6 = recommend_activity(current_facts_6)
print(f"Facts: {current_facts_6} -> Recommendation: {recommendation_6}")
```

### Q&A
The developed expert system uses 'if-then' rules to recommend activities based on weather conditions. Its decision-making capabilities are based on a direct matching of input facts against predefined conditions in the rules.

The system's rules are structured as a list of dictionaries. Each dictionary represents a rule and contains two keys:
1.  `"conditions"`: A dictionary specifying the 'if' part of the rule (e.g., `{'weather': 'sunny', 'temperature': 'warm'}`).
2.  `"recommendation"`: A string describing the 'then' part (e.g., `'Go for a walk or a picnic in the park.'`).

The inference engine, implemented as the `recommend_activity` function, takes a dictionary of `facts` as input. It iterates through the defined rules, checking if all conditions within a rule match the provided facts. If a match is found, it returns the corresponding recommendation. If no rule's conditions are fully met by the given facts, it returns a default message: "No specific recommendation based on current facts."
